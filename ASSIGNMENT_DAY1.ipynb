# QUESTION-2

```python
x = np.arange(1,8) #days of a week 
#EVEN "x" CAN BE WRITTEN LIKE THIS
sales_1 = [160,150,140,145,175,165,180] #sales on a particular days for company 1
sales_2 = [70,90,160,150,140,145,175] #sales on a particular days for company 2
```

```python
x
```

```python
sales_1
```

```python
sales_2
```

```python
plt.plot(x,sales_1,'r*-')
plt.title('LINE PLOT for displaying sales of company 1', fontsize = 16)
plt.xlabel("X-axis", fontsize = 13, color = 'b')
plt.ylabel("Y-axis", fontsize = 13, color = 'b')
```

```python
plt.plot(x,sales_2,'r*-')
plt.title('LINE PLOT for displaying sales of company 2', fontsize = 16)
plt.xlabel("X-axis", fontsize = 13, color = 'b')
plt.ylabel("Y-axis", fontsize = 13, color = 'b')
```

```python
plt.figure(figsize=(6,5), dpi=100)
plt.title('LINE PLOT for displaying sales of company 1 and 2', fontsize = 16)
plt.xlabel("X-axis", fontsize = 13, color = 'b')
plt.ylabel("Y-axis", fontsize = 13, color = 'b')
plt.plot(x,sales_1,'r*-',label = 'sales_1')
plt.plot(x,sales_2,'b.-',label = 'sales_2')
plt.legend(loc='upper left')
plt.grid()
```
