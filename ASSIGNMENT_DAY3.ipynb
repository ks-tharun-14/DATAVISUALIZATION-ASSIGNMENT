###Plot using sns.relplot, line plot, box plot on fmri data set

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
import seaborn as sns
from numpy.random import randn, randint, uniform, sample
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri
```

```python
fmri['region'].unique()
```

```python
sns.relplot(x ='timepoint', y  ='signal', hue = 'event', data = fmri) 
plt.title("FMRI PLOT", fontweight = 'bold')
plt.show()
```

```python
sns.relplot(x ='timepoint', y  ='signal', hue = 'event', style = 'region', data = fmri) 
plt.title("FMRI PLOT", fontweight = 'bold')
plt.show()
```

##LINE PLOT

```python
sns.relplot(data=fmri, x="timepoint", y="signal",hue="event", kind="line")
plt.title("FMRI LINE PLOT", fontweight = 'bold')
plt.show()
```

```python
sns.relplot(x ='timepoint', y  ='signal', hue = 'event', style = 'region', kind = 'line', height=7, aspect = 1.3, data = fmri)
plt.title("FMRI LINE PLOT", fontweight = 'bold')
plt.show()
```

##BOX PLOT

```python
sns.catplot(x ='timepoint', y  ='signal', hue = 'event', kind = 'box', height=9, aspect = 1.4, data = fmri, fliersize = '.5') 
plt.title("FMRI BOX PLOT", fontweight = 'bold')
plt.show()
```

```python
sns.catplot(x ='timepoint', y  ='signal', hue = 'event', col = 'region', kind = 'box', height=7, aspect = 1.3 ,data = fmri)
plt.show()
```
